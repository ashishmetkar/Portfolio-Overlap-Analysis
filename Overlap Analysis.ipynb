```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('/Users/ashishmetkar/Documents/Languages/Projects/Overlap Analysis/Sample Portfolios.xlsx',sheet_name='Sheet1',skiprows=0)
```

```python
#OVERLAP BASIS ISIN
# Function to calculate overlap between two funds
def overlap(fund1, fund2):
    # Get the stocks held by fund1 and fund2
    stocks1 = set(df.loc[df['Scheme Name'] == fund1, 'ISIN'])
    stocks2 = set(df.loc[df['Scheme Name'] == fund2, 'ISIN'])
    # Get the common stocks
    common_stocks = stocks1.intersection(stocks2)
    # Calculate the overlap
    overlap_sum = 0
    for stock in common_stocks:
        weight1 = df.loc[(df['Scheme Name'] == fund1) & (df['ISIN'] == stock), '% of Net Assets']
        weight2 = df.loc[(df['Scheme Name'] == fund2) & (df['ISIN'] == stock), '% of Net Assets']
        if not weight1.empty and not weight2.empty:
            overlap_sum += min(weight1.iloc[0], weight2.iloc[0])
    return overlap_sum


# Function to generate the overlap table
def overlap_table(funds):
    # Initialize the table with zeros
    table = pd.DataFrame(0, index=funds, columns=funds)
    # Calculate the overlap between each pair of funds
    for i in range(len(funds)):
        for j in range(i+1, len(funds)):
            table.iloc[i, j] = overlap(funds[i], funds[j])
            table.iloc[j, i] = table.iloc[i, j]
    return table

funds = ['Kotak Flexicap Fund',
'HDFC Index Fund-NIFTY 50 Plan',
'Canara Robeco Flexi Cap Fund',
'HDFC Flexi Cap Fund',
'ICICI Prudential Flexicap Fund',
'Quant Flexi Cap Fund',
'SBI Flexicap Fund',
'UTI Flexi Cap Fund',
'WhiteOak Capital Flexi Cap Fund',
'Axis Small Cap Fund',
'Mirae Asset Large Cap Fund',
'Mirae Asset Midcap Fund',
'HDFC Large and Mid Cap Fund']
table = overlap_table(funds)
```

```python
#OVERLAP BASIS COMPANY NAME
# Function to calculate overlap between two funds2
def overlap(fund1, fund2):
    # Get the stocks held by fund1 and fund2
    stocks1 = set(df.loc[df['Scheme Name'] == fund1, 'Company Name'])
    stocks2 = set(df.loc[df['Scheme Name'] == fund2, 'Company Name'])
    # Get the common stocks
    common_stocks = stocks1.intersection(stocks2)
    # Calculate the overlap
    overlap_sum = 0
    for stock in common_stocks:
        weight1 = df.loc[(df['Scheme Name'] == fund1) & (df['Company Name'] == stock), '% of Net Assets']
        weight2 = df.loc[(df['Scheme Name'] == fund2) & (df['Company Name'] == stock), '% of Net Assets']
        if not weight1.empty and not weight2.empty:
            overlap_sum += min(weight1.iloc[0], weight2.iloc[0])
    return overlap_sum


# Function to generate the overlap table_comp
def overlap_table_comp(funds2):
    # Initialize the table_comp with zeros
    table_comp = pd.DataFrame(0, index=funds2, columns=funds2)
    # Calculate the overlap between each pair of funds2
    for i in range(len(funds2)):
        for j in range(i+1, len(funds2)):
            table_comp.iloc[i, j] = overlap(funds2[i], funds2[j])
            table_comp.iloc[j, i] = table_comp.iloc[i, j]
    return table_comp

# Example usage
funds2 = ['Kotak Flexicap Fund',
'HDFC Index Fund-NIFTY 50 Plan',
'Canara Robeco Flexi Cap Fund',
'HDFC Flexi Cap Fund',
'ICICI Prudential Flexicap Fund',
'Quant Flexi Cap Fund',
'SBI Flexicap Fund',
'UTI Flexi Cap Fund',
'WhiteOak Capital Flexi Cap Fund',
'Axis Small Cap Fund',
'Mirae Asset Large Cap Fund',
'Mirae Asset Midcap Fund',
'HDFC Large and Mid Cap Fund'
]
table_comp = overlap_table_comp(funds2)
table_comp
```
